### Analisando Dados Cidades Digitais

Dados obtidos do Portal Brasileiro de Dados Abertos: http://dados.gov.br/dataset/cidades-digitais

```python
#Importando bibliotecas
import pandas as pd
import numpy as np
```

```python
#Lendo o arquivo csv
df_cidades_digitais = pd.read_csv('cidades_digitais.csv', sep = ';', encoding = 'latin1')
```

```python
#Visualizando as primeiras linhas do dataframe
df_cidades_digitais.head()
```

```python
#Deletando a coluna IBGE
del df_cidades_digitais['IBGE']
```

```python
#Visualizando se a coluna foi apagada
df_cidades_digitais.head()
```

```python
#Aplicando Group By para agrupar as cidades por UF
cidades_by_uf = df_cidades_digitais.groupby('UF')
```

```python
type(cidades_by_uf)
```

```python
#Visualizando os grupos
cidades_by_uf.groups
```

```python
#Visualizando o número de grupos criados
len(cidades_by_uf.groups)
```

```python
#Visualizando o tamanho de cada grupo e armazenando na variável ordenar_uf
ordenar_uf = cidades_by_uf.size()
ordenar_uf
```

```python
#Ordenando os grupos por ordem crescente
ordenar_uf.sort_values(ascending=False)
```

```python
#Agrupando por UF e STATUS
cidades_by_uf_status = df_cidades_digitais.groupby(['UF', 'STATUS'])
```

```python
#Total agrupado
cidades_by_uf_status.size()
```
